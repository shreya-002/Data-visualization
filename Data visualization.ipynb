# Data Exploration and Cleaning
Data exploration is the first step of data analysis used to explore and visualize data to uncover insights from the start or identify areas or patterns to dig into more. Using interactive dashboards and point-and-click data exploration, users can better understand the bigger picture and get to insights faster.

```python
# Ploting Neighbourhood_group
labels = ['Brooklyn', 'Bronx', 'Queens', 'Manhattan', 'Staten Island']
values = [111, 2, 11, 124, 1]
plt.pie(values, labels = labels, autopct = '%.1f%%')
plt.show()
```

```python
labels = ['Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Bronx']
values = [111, 124, 11, 1, 2]
positions= [0,1,2,3,4]
plt.bar(positions, values, width=0.5)
plt.xticks(positions, labels)
plt.title('Share of neighbourhood')
plt.show()
```

```python
# Checking price of airbnb in neighbourhood_group
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data= data, x = data['neighbourhood_group'], y = data['price'])
```

```python
data['neighbourhood'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(60,15)})
sns.histplot(data['neighbourhood'])
```

```python
labels = ['Entire home/apt', 'Private room', 'Shared room']
values = [134, 113, 2]
explode= [ 0.05, 0, 0]
colors = ['g', 'r', 'b']
plt.figure(figsize=(5,5))
plt.pie(values, labels = labels, autopct = '%.1f%%', explode= explode, colors= colors)
plt.show()
```

```python
# Analysing number of reviews in naighbourhood group
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data= data, x = data['neighbourhood_group'], y = data['number_of_reviews'])
```
